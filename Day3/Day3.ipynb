```python
import pandas as pd
import numpy as np
data = pd.read_csv('inputPart1.txt', header=None)
data.head()
input = data.iloc[:,:].values
schematic = ["".join(row) for row in input]
symbols = ["*", "+", "$", "#", "/", "=", "%", "@", "-", "&"]
```

```python
def is_adjacent_symbol(schematic, row, col, symbols):
    character = schematic[row][col]
    for dx in range(-1, 2):
        for dy in range(-1, 2):
            if dx == 0 and dy == 0:
                continue
            neghbour_row = row + dx
            neghbour_col = col + dy
            if neghbour_row < 0 or neghbour_row >= len(schematic):
                continue
            if neghbour_col < 0 or neghbour_col >= len(schematic[0]):
                continue
            if schematic[neghbour_row][neghbour_col] in symbols and character.isdigit():
                return (character if character.isdigit() else "."), True 
    return (character if character.isdigit() else "."), False

def extract_numbers(schematic):
    numbers = []

    for row in schematic:
        current_number = ""
        for char in row:
            if char.isdigit():
                current_number += char
            elif current_number:
                numbers.append(int(current_number))
                current_number = ""

        if current_number:
            numbers.append(int(current_number))

    return numbers
```

```python
import itertools
import math
import re
from collections import defaultdict

with open("inputPart1.txt") as f:
    ls = f.read().strip().split("\n")

box = list(itertools.product((-1, 0, 1), (-1, 0, 1)))
symbols = {
    (i, j)
    for i, l in enumerate(ls)
    for j, x in enumerate(l)
    if x != "." and not x.isdigit()
}

part_sum = 0
parts_by_symbol = defaultdict(list)

result = []
for i, l in enumerate(ls):
    for match in re.finditer(r"\d+", l):
        n = int(match.group(0))
        boundary = {
            (i + di, j + dj)
            for di, dj in box
            for j in range(match.start(), match.end())
        }
        if symbols & boundary:
            result.append(n)
            part_sum += n
        for symbol in symbols & boundary:
            parts_by_symbol[symbol].append(n)
print(parts_by_symbol.values())
print(result)
print(part_sum)
```
